```python
%%jamcell test --port COM5 --board attinyx5opti

// the setup function runs once when you press reset or power the board
void setup() {
  // initialize digital pin LED_BUILTIN as an output.
  pinMode(2, OUTPUT);
}

// the loop function runs over and over again forever
void loop() {
  digitalWrite(2, HIGH);   // turn the LED on (HIGH is the voltage level)
  delay(500);                       // wait for a second
  digitalWrite(2, LOW);    // turn the LED off by making the voltage LOW
  delay(500);                       // wait for a second
}
```

```python
%%jamcell test --port COM5 --board attinyx5opti

#include <Adafruit_NeoPixel.h>

Adafruit_NeoPixel pixels(9, 3, NEO_GRB + NEO_KHZ800);

void SimpleFlashPixels(int r, int g, int b){

   pixels.clear(); // Set all pixel colors to 'off'

  // The first NeoPixel in a strand is #0, second is 1, all the way up
  // to the count of pixels minus one.
  for(int i=0; i<9; i++) { // For each pixel...

    // pixels.Color() takes RGB values, from 0,0,0 up to 255,255,255
    // Here we're using a moderately bright green color:
    pixels.setPixelColor(i, pixels.Color(r, g, b));

    pixels.show();   // Send the updated pixel colors to the hardware.

    delay(200); // Pause before next pass through loop
  }}

void setup() {
    pixels.begin();
    SimpleFlashPixels(0,255,0);
    
}

void loop() {

  SimpleFlashPixels(255,0,255);
  SimpleFlashPixels(255,0,0);
  SimpleFlashPixels(0,0,255);
    
}
```

```python
%%jamcell test --port COM5 --board attinyx5opti --plot 200

// Define the analog pin, the TMP36's Vout pin is connected to
#define sensorPin A1

void setup() {
  // Begin serial communication at 9600 baud rate
  Serial.begin(9600);
}

void loop() {
  // Get the voltage reading from the TMP36
  int reading = analogRead(sensorPin);

  // Convert that reading into voltage
  // Replace 5.0 with 3.3, if you are using a 3.3V Arduino
  float voltage = reading * (5.0 / 1024.0);

  // Convert the voltage into the temperature in Celsius
  float temperatureC = (voltage - 0.5) * 100;

  // Print the temperature in Celsius
  
  Serial.print(temperatureC);
  Serial.print(" ");
  
  // Print the temperature in Fahrenheit
  float temperatureF = (temperatureC * 9.0 / 5.0) + 32.0;
  Serial.println(temperatureF);
 

  delay(1000); // wait a second between readings
}
```
